```python
from scipy.stats import norm

s_norm = norm().rvs(10000).reshape(-1, 1)
s_over = norm(0, 2).rvs(10000).reshape(-1, 1)
s_under = norm(0, 0.7).rvs(10000).reshape(-1, 1)

from valdiags.plot_utils import multi_corner_plots

multi_corner_plots(
    [s_norm, s_over, s_under],
    legends=[r"$q(\theta\mid x)$", r"$p_1(\theta\mid x)$", r"$p_2(\theta\mid x)$"],
    colors=["blue", "green", "red"],
    title="PDFs",
    labels=[r'$\theta$'],
    domain=([-5],[5]),
    creds=[]

)
import matplotlib.pyplot as plt
plt.savefig('pdfs_accuracy.pdf')
```
